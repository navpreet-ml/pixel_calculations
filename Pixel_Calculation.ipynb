```python
# importing libraries
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# reading the image data
img_before = cv2.imread("Minor bottom clean process.jpg", cv2.IMREAD_GRAYSCALE)
thresh = 1
img_after = cv2.threshold(img_before, thresh, 255, cv2.THRESH_BINARY)[1]
#Show the image with matplotlib
plt.imshow(img_before, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
print(f"Total number of pixels in image before = {str(img_before.shape[0]*img_before.shape[1])}")
```

```python
# reading the image data 
img_after = cv2.imread("Minor bottom after process.jpg", cv2.IMREAD_GRAYSCALE)
thresh = 1
img_after = cv2.threshold(img_after, thresh, 255, cv2.THRESH_BINARY)[1]
#Show the image with matplotlib
plt.imshow(img_after, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
print(f"Total number of pixels in image before = {str(img_after.shape[0]*img_after.shape[1])}")
```

```python
# counting the number of pixels
number_of_white_pix_before = np.sum(img_before > 0)
number_of_black_pix_before = np.sum(img_before == 0)
```

```python
# counting the number of pixels
number_of_white_pix_after = np.sum(img_after > 0)
number_of_black_pix_after = np.sum(img_after == 0)
```

```python
print('Number of white pixels before:', number_of_white_pix_before)
print('Number of black pixels before:', number_of_black_pix_before)
```

```python
print('Number of white pixels after:', number_of_white_pix_after)
print('Number of black pixels after:', number_of_black_pix_after)
```
